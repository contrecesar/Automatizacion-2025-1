```python
import pandas as pd
df =pd.read_csv('./data/coffee_data.csv')

city_satisfaction = df.groupby("city")["satisfaction_score"].mean()

# Sort cities by satisfaction score in descending order
city_satisfaction_sorted = city_satisfaction.sort_values(ascending=False).reset_index()

# Display the top cities
print("Cities ranked by average satisfaction score:\n")
print(city_satisfaction_sorted)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
city_satisfaction.plot(kind='bar', color='skyblue')
plt.title("Average Customer Satisfaction by City")
plt.xlabel("City")
plt.ylabel("Average Satisfaction Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
ny_data = df[df["city"] == "New York"]

# Group by product and sum quantities sold
product_sales_ny = ny_data.groupby("product")["units_sold"].sum().sort_values()

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
product_sales_ny.plot(kind='barh', color='peru')
plt.title("Total Product Sales in New York")
plt.xlabel("Quantity Sold")
plt.ylabel("Product")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="satisfaction_score", bins=10, kde=True, color='skyblue')
plt.title("Distribution of Satisfaction Scores")
plt.xlabel("Satisfaction Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of units sold vs satisfaction score, colored by city
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="units_sold", y="satisfaction_score", hue="city", palette="deep")
plt.title("Relationship Between Units Sold and Satisfaction Score")
plt.xlabel("Units Sold")
plt.ylabel("Satisfaction Score")
plt.legend(title="City", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
